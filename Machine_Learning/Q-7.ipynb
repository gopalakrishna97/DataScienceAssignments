```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('data.csv')
data
```

```python
data.info()
```

```python
data['label'].value_counts()
```

```python
label_encoder = LabelEncoder()
data['label'] = label_encoder.fit_transform(data["label"])
```

```python
label_encoder.inverse_transform(data['label'])
```

```python
data
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
```

```python
X = data.drop('filename', axis=1)

X = X.loc[:, X.columns != 'label']
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
# Cluster labels
labels = kmeans.labels_
data["cluster"] = labels
labels
```

```python
# Silhouette score
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)
```

```python
import seaborn as sns
data["cluster"] = kmeans.labels_
sns.lmplot(x='tempo', y='spectral_centroid', data=data, hue='cluster', palette='coolwarm', aspect=1, fit_reg=False)
```

```python
import pickle

# Save the model to a file
with open('music_feature_model.pkl', 'wb') as f:
    pickle.dump(kmeans, f)
```

```python

# Save the labelEncoder to a file
with open('labelencoder.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)
```
