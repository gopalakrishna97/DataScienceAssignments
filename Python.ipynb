## Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

`Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5`

```python
from collections import Counter
def maxLenWordOfMostFrequentlyOccuredWords(string:str):
    freq = Counter(string.split()).most_common()
    heighest_frequency_of_word = freq[0][1]
    most_frequent_words = filter(lambda x : x[1]>=heighest_frequency_of_word ,freq)
    maximum_valued_word = ""
    for i in most_frequent_words:
        if len(i[0])>len(str(maximum_valued_word)):
            maximum_valued_word = i[0]
    print(f"most frequently occured maximum length word is '{maximum_valued_word}' it's length is {len(maximum_valued_word)} and it's occurances are {heighest_frequency_of_word}")
    
maxLenWordOfMostFrequentlyOccuredWords("Learn and practice and learn to practice")
maxLenWordOfMostFrequentlyOccuredWords("write write write all the number from from from 1 to 100")
```

## Question 2: -
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

`Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO`

```python
from collections import Counter
def isvalid_dict(string:str):
    res = Counter(string)
    occurances = list(res.values())
    result = all(element == occurances[0] for element in occurances)
    if(result):
        print('YES')
    else:
        print('NO')

isvalid_dict("abc")
isvalid_dict("abcc")
```

## Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
nltk.download()
```

```python


from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize

# Function to extract the parts of speech 

def partsOfSpeech(text):
    
    dic = {"nouns": 0,
           "pronouns": 0,
           "verbs": 0,
           "adjectives": 0
          }
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        words = [word for word in words if word not in set(stopwords.words('english'))]
        tagged = nltk.pos_tag(words)
        for (word, tag) in tagged:
            if tag == 'NNP': # If the word is a proper noun

                dic["nouns"] += 1
            elif tag in ['PRP','WP']:
                dic['pronouns'] += 1
                
            elif tag in ['VB','VBD','VBG','VBN','VBP','VBZ']:
                dic['verbs'] += 1
                
            elif tag in ['JJ','JJR','JJS']:
                dic['adjectives'] += 1
    print(dic)
text1 =  "Rohan is a wonderful player. He was born in India. He is a fan of the movie Wolverine. He has a dog named Bruno."
text2 = "A cherry is the fruit of many plants of the genus Prunus, and is a fleshy drupe"
text3 = "We are an impact-driven global non-profit working on the ground to create scalable solutions for nature and people to thrive. In the field, we create, test, and scale new ways of production and doing business."

partsOfSpeech(text1)
partsOfSpeech(text2)
partsOfSpeech(text3)
```
